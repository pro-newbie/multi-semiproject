```python
import pandas as pd
from sklearn import preprocessing

filename = 'DATA_2022년_국민여행조사_원자료_M.csv'

df = pd.read_csv(filename, encoding='cp949')
```

```python
df=df.dropna()
```

```python
import pandas as pd
import numpy as np

bins = [-1, 1, 100000, 500000, 1000000, 2000000, 3000000, np.inf]
labels = [0, 1, 2, 3, 4, 5, 6]
df['A9F'] = pd.cut(df['A9F'], bins=bins, labels=labels)
df['A9F']
```

```python
import pandas as pd
import numpy as np

bins = [-1, 1, 50000, 100000, 3000000, 5000000, 7000000, np.inf]
labels = [0, 1, 2, 3, 4, 5, 6]
df['NA9C'] = pd.cut(df['NA9C'], bins=bins, labels=labels)
df['NA9C']
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

encoder = LabelEncoder()
df = df.apply(encoder.fit_transform)
df
```

```python
df = df.rename(columns={'국내_숙박여부': 'Accommodation',
                       'A9C' : 'Total accommodation fee',
                       'A9F': 'Total transportation cost',
                      'D_TRA1_1_SPOT':'SPOT',
                        'BAGE':'AGE',
                        'BSEX':'SEX',
                        'A7': 'Number_of_people'})
df 
```

```python
df=df[['SPOT', 'AGE', 'SEX', 'Total accommodation fee', 'Total transportation cost', 'Number_of_people', 'Accommodation']]
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr = df.corr()


plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
def convert_number_of_people(value):
    if value in [3, 4, 5, 6, 7, 8, 9]:
        return 3
    elif value >= 10:
        return 4
    else:
        return value

df['Number_of_people'] =df['Number_of_people'].apply(convert_number_of_people)

df
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
```

```python
X = df.drop('Accommodation', axis=1)
y = df['Accommodation']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = svm.SVC()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)


report = metrics.classification_report(y_test, y_pred)
print("Report: \n", report)
```

```python
#교통비, 동행인수 , 숙박여부만 남김

df = df.drop('SPOT', axis = 1)
df = df.drop('SEX', axis = 1)
df = df.drop('AGE', axis = 1)
df = df.drop('Total accommodation fee', axis = 1)
df
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, accuracy_score

# 데이터 분할
Y_data = df['Accommodation']
X_data = df.drop('Accommodation', axis=1)
X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_data, test_size=0.2, random_state=42)
```

```python
 
svc = svm.SVC(kernel='rbf')
parameters = {
    'C': [0.1, 1, 10, 100],  
    'gamma': ['scale', 'auto', 0.001, 0.01, 0.1, 1]  
}

# 그리드 서치 
grid_search = GridSearchCV(svc, parameters, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, Y_train)

# 최적의 하이퍼파라미터 
best_parameters = grid_search.best_params_
print("Best Parameters:", best_parameters)

best_clf = grid_search.best_estimator_
predictions = best_clf.predict(X_test)

accuracy = accuracy_score(Y_test, predictions)
print(f'Accuracy: {accuracy*100:.2f}%')

print(classification_report(Y_test, predictions))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm


best_clf = grid_search.best_estimator_
x_min, x_max = X_train.iloc[:, 0].min() - 1, X_train.iloc[:, 0].max() + 1
y_min, y_max = X_train.iloc[:, 1].min() - 1, X_train.iloc[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

Z = best_clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)
plt.scatter(X_train.iloc[:, 0], X_train.iloc[:, 1], c=Y_train, s=20, edgecolor='k', cmap=plt.cm.coolwarm)
plt.xlabel('Total transportation cost')
plt.ylabel('Number_of_people')
plt.title('SVM Decision Boundary with Best Parameters')
plt.xlim(left=0)
plt.ylim(bottom=0)

plt.show()
```
